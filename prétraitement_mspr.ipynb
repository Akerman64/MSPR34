```python
import pandas as pd

# Spécifier les chemins des 4 fichiers CSV
fichiers = [
    "eco2mix_2018.csv",
    "eco2mix_2019.csv",
    "eco2mix_2020.csv",
    "eco2mix_2021.csv"
]

# Lire chaque fichier CSV et les ajouter à une liste
dfs = [pd.read_csv(fichier) for fichier in fichiers]

# Fusionner tous les DataFrames dans la liste (par concaténation par défaut)
df_fusionne = pd.concat(dfs, ignore_index=True)

# Afficher les premières lignes du DataFrame fusionné
print(df_fusionne.head())

# Sauvegarder le DataFrame fusionné dans un nouveau fichier CSV
df_fusionne.to_csv("fichier_fusionne.csv", index=False, encoding='utf-8')
```

```python
import pandas as pd

# Charger le fichier CSV
df = pd.read_csv("eco2mix2018-2019_pour_entrainement.csv")

# S'assurer que la colonne 'Date' est bien en format datetime
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Supprimer les lignes où la date est manquante (ex: la dernière ligne vide)
df = df.dropna(subset=['Date'])

# Grouper par jour et calculer la moyenne des consommations
df_journalier = df.groupby(df['Date'].dt.date).agg({
    'Consommation': 'mean'
}).reset_index()

# Renommer les colonnes pour plus de clarté
df_journalier.columns = ['Date', 'Consommation']

# Sauvegarder le résultat si besoin
df_journalier.to_csv("consommation_journaliere.csv", index=False)

print(df_journalier.head())
```

```python
import pandas as pd

# Charger les deux datasets
df1 = pd.read_csv("consommation_journaliere.csv")
df2 = pd.read_csv("first_fusion.csv")  # Remplacez par le chemin du deuxième fichier

# S'assurer que la colonne 'Date' est bien en format datetime pour les deux datasets
df1['Date'] = pd.to_datetime(df1['Date'], errors='coerce')
df2['Date'] = pd.to_datetime(df2['Date'], errors='coerce')

# Fusionner les deux datasets selon la colonne 'Date'
df_fusionne = pd.merge(df1, df2, on='Date', how='inner')  # Utilisez 'left', 'right', ou 'outer' selon vos besoins

# Sauvegarder le fichier fusionné si besoin
df_fusionne.to_csv("eco2mix_final.csv", index=False)

# Afficher les premières lignes du dataset fusionné
print(df_fusionne.head())
```
